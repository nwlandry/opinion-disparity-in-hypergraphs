```python
import json
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters
```

```python
with open("Data/stability/mean-field_beta2_beta3_polarization.json") as file:
    data = json.load(file)
```

```python
beta2 = np.array(data["beta2"], dtype=float)
beta3 = np.array(data["beta3"], dtype=float)
psi = np.array(data["psi"], dtype=float)
```

```python
# define "next to" - this may be a square, diamond, etc
# selem = skimage.morphology.diamond(1)
# black_white_edges = (skimage.filters.rank.minimum(img, selem) == 0) & (skimage.filters.rank.maximum(img, selem) == 1)
# # plt.imshow(black_white_edges)
# black_white_edges
plt.figure()
for i in range(5):
    plt.subplot(1,5,i+1)
    edge_roberts = filters.roberts(np.flipud(psi))
    edge_roberts[edge_roberts > 0.1] = i*0.1+0.4
    edge_roberts[edge_roberts < 0.1] = 0
    plt.imshow(edge_roberts, cmap="Blues", vmin=0, vmax=1, extent=(min(beta2), max(beta2), min(beta3), max(beta3)), aspect="auto")
```
