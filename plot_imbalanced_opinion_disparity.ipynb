```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
tick_label_fontsize = 12
axis_label_fontsize = 20
legend_fontsize = 14
```

```python
with open("Data/opiniondisparity/mean-field_rho_epsilon2_opinion_disparity.json") as file:
    data = json.load(file)

rho = np.array(data["rho"], dtype=float)
epsilon2 = np.array(data["epsilon2"], dtype=float)
psi1 = np.array(data["psi1"], dtype=float)
psi2 = np.array(data["psi2"], dtype=float)
```

```python
xsec_index1 = 50
xsec_index2 = 18
rho1 = 0.506
rho2 = 0.515

print(epsilon2[xsec_index1])
print(epsilon2[xsec_index2])


fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(6, 8), height_ratios=[1, 10, 3, 3])

axes[0].tick_params(axis="x", labelsize=tick_label_fontsize)

im = axes[1].imshow(
    np.flipud(psi1 - psi2),
    extent=[min(rho), max(rho), min(epsilon2), max(epsilon2)],
    aspect="auto",
    cmap="inferno",
)
axes[1].plot([0.5, 0.5], [min(epsilon2), max(epsilon2)], "k--")
axes[1].plot(
    [min(rho), max(rho)],
    [epsilon2[xsec_index1], epsilon2[xsec_index1]],
    "-",
    color="darkgrey",
)
axes[1].plot(
    [min(rho), max(rho)],
    [epsilon2[xsec_index2], epsilon2[xsec_index2]],
    "-",
    color="teal",
)
axes[1].scatter(
    rho1,
    epsilon2[xsec_index1],
    color="darkgrey",
    edgecolors="black",
    s=100,
    marker="*",
    zorder=2,
)
axes[1].scatter(
    rho2,
    epsilon2[xsec_index1],
    color="darkgrey",
    edgecolors="black",
    s=100,
    marker="*",
    zorder=2,
)
axes[1].scatter(
    rho1,
    epsilon2[xsec_index2],
    color="teal",
    edgecolors="white",
    s=100,
    marker="*",
    zorder=2,
)
axes[1].scatter(
    rho2,
    epsilon2[xsec_index2],
    color="teal",
    edgecolors="white",
    s=100,
    marker="*",
    zorder=2,
)

axes[1].set_xticks(
    [0.48, 0.49, 0.5, 0.51, 0.52],
    [0.48, 0.49, 0.5, 0.51, 0.52],
    fontsize=tick_label_fontsize,
)
axes[1].set_yticks([0.0, 0.5, 1], fontsize=tick_label_fontsize)
axes[1].set_ylabel(r"$\epsilon_2$", fontsize=16)

axes[2].plot(rho, psi1[xsec_index1, :], label=r"$\psi_1$", color="#1A85FF")
axes[2].plot(rho, psi2[xsec_index1, :], label=r"$\psi_2$", color="#D41159")
axes[2].plot([0.5, 0.5], [0, 0.6], "k--")
axes[2].set_xlim([min(rho), max(rho)])
axes[2].set_ylim([0, 0.6])
axes[2].set_yticks([0, 0.25, 0.5], [0.0, 0.25, 0.5], fontsize=tick_label_fontsize)
axes[2].set_xticks(
    [0.48, 0.49, 0.5, 0.51, 0.52],
    [0.48, 0.49, 0.5, 0.51, 0.52],
    fontsize=tick_label_fontsize,
)

axes[3].plot(rho, psi1[xsec_index2, :], label=r"$\psi_1$", color="#1A85FF")
axes[3].plot(rho, psi2[xsec_index2, :], label=r"$\psi_2$", color="#D41159")
axes[3].plot([0.5, 0.5], [0, 0.6], "k--")
axes[3].set_xlim([min(rho), max(rho)])
axes[3].set_ylim([0, 0.6])
axes[3].set_xticks(
    [0.48, 0.49, 0.5, 0.51, 0.52],
    [0.48, 0.49, 0.5, 0.51, 0.52],
    fontsize=tick_label_fontsize,
)
axes[3].set_yticks([0, 0.25, 0.5], [0.0, 0.25, 0.5], fontsize=tick_label_fontsize)
axes[3].legend(fontsize=legend_fontsize)
axes[3].set_xlabel(r"$\rho$", fontsize=axis_label_fontsize)

# add text

# Axis 1
## Line labels
axes[1].text(0.485, 0.475, "(1)", fontsize=16, color="darkgrey")
axes[1].text(0.485, 0.155, "(2)", fontsize=16, color="teal")

# Marker labels
axes[1].text(0.5048, 0.54, "(a)", fontsize=16, color="black")
axes[1].text(0.5138, 0.54, "(b)", fontsize=16, color="black")
axes[1].text(0.5048, 0.22, "(c)", fontsize=16, color="white")
axes[1].text(0.5138, 0.22, "(d)", fontsize=16, color="black")

# fig 2 subplot labels
axes[2].text(0.481, 0.4, "(1)", fontsize=16)

# fig 3 subplot labels
axes[3].text(0.481, 0.4, "(2)", fontsize=16)

cbar = fig.colorbar(im, cax=axes[0], orientation="horizontal", pad=0.01)
cbar.ax.set_title(r"$\psi_1 - \psi_2$", fontsize=axis_label_fontsize)

plt.tight_layout()
plt.savefig("Figures/Fig7/mean-field_rho_epsilon2_psi.png", dpi=1000)
plt.savefig("Figures/Fig7/mean-field_rho_epsilon2_psi.pdf", dpi=1000)
plt.show()
```
