```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("Data/stability/mean-field_epsilon2_epsilon3_polarization.json") as file:
    data1 = json.load(file)
with open("Data/stability/mean-field_beta2_beta3_polarization.json") as file:
    data2 = json.load(file)
with open("Data/stability/mean-field_max_polarization.json") as file:
    data3 = json.load(file)
with open("Data/stability/mean-field_rho_epsilon2_polarization.json") as file:
    data4 = json.load(file)
```

```python
epsilon2 = np.array(data1["epsilon2"], dtype=float)
epsilon3 = np.array(data1["epsilon3"], dtype=float)
psi = np.array(data1["psi"], dtype=float)
```

```python
plt.figure(figsize=(8, 6))
im = plt.imshow(np.flipud(psi), extent=[min(epsilon2), max(epsilon2), min(epsilon3), max(epsilon3)], vmin=0, vmax=1, aspect="auto", cmap="inferno")
plt.xlabel(r"$\epsilon_2$", fontsize=36)
plt.ylabel(r"$\epsilon_3$", fontsize=36)
plt.xticks([0, 0.5, 1], fontsize=24)
plt.yticks([0.8, 0.9, 1], fontsize=24)
cbar = plt.colorbar(im)
cbar.ax.tick_params(labelsize=18)
cbar.set_label(r"$\psi$", fontsize=36, rotation=270, labelpad=40)

plt.tight_layout()
plt.savefig("Figures/epsilon2_epsilon3_psi_plot.png", dpi=1000)
plt.savefig("Figures/epsilon2_epsilon3_psi_plot.pdf", dpi=1000)
plt.show()
```

```python
rho = np.array(data4["rho"], dtype=float)
epsilon2 = np.array(data4["epsilon2"], dtype=float)
psi1 = np.array(data4["psi1"], dtype=float)
psi2 = np.array(data4["psi2"], dtype=float)
```

```python
xsec_index = 50

plt.figure(figsize=(8, 6))
plt.subplot(211)
im = plt.imshow(np.flipud(psi1-psi2), extent=[min(rho), max(rho), min(epsilon2), max(epsilon2)], aspect="auto", cmap="inferno")
plt.plot([0.5, 0.5], [min(epsilon2), max(epsilon2)], 'k--')
plt.plot([min(rho), max(rho)], [epsilon2[xsec_index], epsilon2[xsec_index]], 'r-')

plt.xlabel(r"$\rho$", fontsize=16)
plt.ylabel(r"$\epsilon_2$", fontsize=16)

cbar = plt.colorbar(im)
cbar.ax.tick_params(labelsize=12)
cbar.set_label(r"$\psi_1 - \psi_2$", fontsize=16, rotation=270, labelpad=20)

plt.subplot(212)
plt.plot(rho, psi1[xsec_index,:], label=r"$\psi_1$")
plt.plot(rho, psi2[xsec_index,:], label=r"$\psi_2$")
plt.plot([0.5, 0.5], [0, 0.6], 'k--')
plt.xlim([min(rho), max(rho)])
plt.legend()

plt.tight_layout()
plt.savefig("Figures/rho_epsilon2_psi_plot.png", dpi=1000)
plt.savefig("Figures/rho_epsilon2_psi_plot.pdf", dpi=1000)
plt.show()
```
